```python
from sklearn import metrics
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import neural_network
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate synthetic dataset
X, y = datasets.make_classification(n_samples=100, n_features=10, n_classes=2)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Plot the first two features of the training set
plt.scatter(X_train[:, 0], X_train[:, 1])
plt.title("Training Data Visualization")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Define and train the MLP classifier
clf = neural_network.MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=200)
clf.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = clf.predict(X_test)

# Print classification report, accuracy, and confusion matrix
print("Classification Report:")
print(metrics.classification_report(y_test, y_pred))

accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}".format(accuracy))

confusion = metrics.confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(confusion)
```
