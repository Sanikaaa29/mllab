```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/LinearRegration.csv')


x = data[['Rain in cm']]
y = data['Month']

model = LinearRegression()
model.fit(x, y)

print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)

print("Prediction for input 200:", model.predict([[200]]))
```

```python

plt.xlabel('Rain in cm')
plt.ylabel('Month')
plt.title('Linear Regression')
plt.scatter(x,y)
plt.show()
```
