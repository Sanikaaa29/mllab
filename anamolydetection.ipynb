```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Generate synthetic data
X = np.random.randn(100, 2)  # 100 random points
X = np.vstack([X, np.array([[5, 5], [6, 6], [7, 7]])])  # Add some anomalies

# Apply DBSCAN for anomaly detection
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X)

# Anomalies are labeled as -1
anomalies = X[labels == -1]

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(anomalies[:, 0], anomalies[:, 1], c='red', marker='x', label='Anomalies')
plt.legend()
plt.title('Anomaly Detection with DBSCAN')
plt.show()

# Print identified anomalies
print("Identified Anomalies:")
print(anomalies)
```
